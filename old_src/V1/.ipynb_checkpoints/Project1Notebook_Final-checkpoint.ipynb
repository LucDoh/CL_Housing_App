# Part 1: Data scraping and preparation

## Step 1: Scraping
## Here we import the relevant packages and then use requests to do a get request on spaceweatherlive.com of the top 50 solar flares. We then use BeautifulSoup to parse out the table from the html, and store this into the Pandas Dataframe df. This is then printed without the indices. 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
r = requests.get('https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares')
r.text;
soup = BeautifulSoup(r.content, 'lxml');
soup.prettify();
```

```python
name_box = soup.find('table', attrs={'class': 'table table-striped table-responsive-md'})#.find_all('td') 
table = soup.find('table', attrs={'class': 'table table-striped table-responsive-md'})
```

```python
df = pd.read_html(str(table))
cols = ['rank', 'x_class', 'date', 'region', 'start_time', 'max_time', 'end_time', 'movie']
df[0].columns = cols

print(df[0].shape)

print(df[0].to_string(index=False))

#for i in name_box.find('tbody').find_all('td'):
#    print(i.text + '%%%')
```

## Step 2: Tidying
## Here we tidy the data as required. We combine the date and time columns for start, end and max, into single datetime columns through the function makeDateTime was used to parse these, and then the datetime package was used to combine these into datetime objects. We also drop the unneeded columns and set missing values to NaN.

```python
#Step 2: TIDYING 
import datetime
import time
#Drop the last column:
df[0] = df[0].drop(columns='movie')

#s1 = "2000/07/14" s3 = "19:29" #starttime_date = s1_list + s3
#Need to create start_datetime  max_datetime  end_datetime

def makeDateTime(s1,s3):
    s3_list = list(s3.split(":"))
    s1_list = list(s1.split("/"))
    d = datetime.datetime.combine(datetime.date(int(s1_list[0]),int(s1_list[1]),int(s1_list[2])), 
                                  datetime.time(int(s3_list[0]), int(s3_list[1])))
    return d

#print(makeDateTime(s1,s3))

#Iterate thru rows and combine date and times:
for i,row in df[0].iterrows():
    s_t = str(row['start_time'])
    s_d = str(row['date'])
    s_et = str(row['end_time'])
    s_max = str(row['max_time'])
    d_start = makeDateTime(s_d, s_t)
    d_end = makeDateTime(s_d, s_et)
    d_max = makeDateTime(s_d, s_max)
    df[0].set_value(i,'start_time', str(d_start))
    df[0].set_value(i,'end_time', str(d_end))
    df[0].set_value(i,'max_time', str(d_max))
#print(df[0].head(5))
    
    
#Rename columns: 
#start_datetime  max_datetime  end_datetime
cols_new = ['rank', 'x_class', 'date', 'region', 'start_datetime', 'max_datetime', 'end_datetime']
df[0].columns = cols_new
df[0]

#Rearrange columns:
cols = df[0].columns.tolist()
cols_rearr = ['rank', 'x_class', 'date', 'start_datetime', 'max_datetime', 'end_datetime', 'region']
df[0] = df[0][cols_rearr]

df[0] = df[0].drop(columns='date') #drop column of dates
df_SWL = df[0]
print(df_SWL)
```

## Step 3: Scrap NASA data
## Here we scrape the data from the nasa site, similarly as before, neglecting the irrelevant lines of text and using pythons string parsing functions to make a 2D list of the solar flare data. We use BeautifulSoup to get the text out of the 'a'  tags. Finally, we import this data into a pandas Dataframe df2, keeping only the important first 14 elements of each line. 

We replace all non-values of the various forms with NaNs and save this to a new dataframe df_nan.

```python
##### STEP 3 ####

site_Nasa = 'https://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html'
r2 = requests.get(site_Nasa)
#####################################
soup2 = BeautifulSoup(r2.content, 'lxml')
list_lineSep = []

#Read the html line by line then throw away some 
for line in r2.text.splitlines():
    list_lineSep.append(line)

list_relLines = list_lineSep[15:-3] # throw away first 14 lines, and last 3.
rows_Cleaned = []
#Take text out of <a> tags
for i in range(len(list_relLines)):
    soupL = BeautifulSoup(list_relLines[i], 'lxml')
    for s in soupL.find_all('a'):
        s.replace_with(s.text)
    rows_Cleaned.append(str(soupL.text))
    
#print(rows_Cleaned)

data = [r.split() for r in rows_Cleaned] # Data as required by Step 3 
print(len(data))
#Put this in a data frame then need to make column names
df = pd.DataFrame(data)

df2 = df.iloc[:,:15].copy()
print(df2)
#### END OF STEP 3 ####


    
    
    
    
  
```

## Step 4: Tidy NASA table
## To tidy this table, we replace all non-values of the various forms with NaNs and save this to a new dataframe df_nan. We then create a column of booleans called isHalo, making the value true if the corresponding row is of a halo flare.  We create another column width_lower_ bound of booleans, filled with True when row's width column contains a lower bound. Finally, we combine date and time columns as we did previously.

```python
###### BEGINNING of STEP 4 #####################
#Replace missing values
df_nan = df2.replace('Back', np.nan)
df_nan = df_nan.replace('---', np.nan)
df_nan = df_nan.replace('-----',np.nan)
df_nan = df_nan.replace('----',np.nan)
df_nan = df_nan.replace('------',np.nan)
for i,row in df_nan.iterrows():
    if df_nan.iloc[i, 2] == '--/--':
        df_nan.iloc[i, 2] = row['cme_date']
        df_nan.iloc[i, 3] = '00:00'        
df_nan = df_nan.replace('--/--', np.nan)
df_nan = df_nan.replace('--:--', np.nan) # Replaced with 23:59am
df_nan = df_nan.replace('24:00', '23:59') # Replaced with 23:59am

#--/--    --:--
#Make isHalo column + fix missing values
isHalo = [False for i in range(df_nan.shape[0])]
for i in range(len(isHalo)):
    if(df_nan.iloc[i,11] == 'Halo'):
        isHalo[i] = True
        
df_nan['is_halo'] = isHalo
df_nan = df_nan.replace('Halo',np.nan)
widthLBound = [False for i in range(df_nan.shape[0])]
for i in range(len(widthLBound)):
    if(str(df_nan.iloc[i,12])[0] == '>'):
        widthLBound[i] = True

df_nan['width_lower_bound'] = widthLBound
df_nan.columns = ['start_date', 'start_time', 'end_date', 'end_time', 'start_frequency', 'end_frequency',
                      'flare_location', 'flare_region', 'importance', 'cme_date','cme_time', 'cme_angle',
                      'cme_width','cme_speed', 'plot','is_halo', 'width_lower_bound']




#Now, combine date and time columns for start, end and cme
def makeDateTime2(s1,s3,y):
    s3_list = list(s3.split(":"))
    s1_list = list(s1.split("/"))
    try:
        d = datetime.datetime.combine(datetime.date(int(y),int(s1_list[0]),int(s1_list[1])), 
                                  datetime.time(int(s3_list[0]), int(s3_list[1])))
    except:
        d = False
    return d


#print(df_nan.iloc[69]) #print(df_nan.iloc[96]) #Problematics times/dates

#Iterate thru rows 
for i,row in df_nan.iterrows():
    s_t = str(row['start_time'])
    s_d = str(row['start_date'])
    d_start = makeDateTime(s_d, s_t)
    y = d_start.year # Save start year so that it can be used for others  
    s_et = str(row['end_time'])
    s_ed = str(row['end_date'])
    s_cmeD = str(row['cme_date'])
    s_cmeT = str(row['cme_time'])
    
    
    d_end = makeDateTime2(s_ed, s_et, y)
    d_cme = makeDateTime2(s_cmeD,s_cmeT,y)
    
    df_nan.set_value(i,'start_time', str(d_start))
    df_nan.set_value(i,'end_time', str(d_end))
    if(d_cme == False):
        df_nan.set_value(i, 'cme_time', str(np.nan))
    else:
        df_nan.set_value(i,'cme_time', str(d_cme))

df_nan.columns = ['start_date', 'start_datetime', 'end_date', 'end_datetime', 'start_frequency', 'end_frequency',
                      'flare_location', 'flare_region', 'importance', 'cme_date','cme_datetime', 'cpa',
                      'width','speed', 'plot', 'is_halo', 'width_lower_bound']
df_nan = df_nan.drop(columns=['start_date', 'end_date', 'cme_date']) 

print(df_nan)

#####END OF STEP 4###############
    
```

Initial idea:

## Question 3: Analysis
Plot one plot that shows the top 50 solar flares in context with all data available in the NASA dataset.

I am going to plot the starting frequencies of the NASA dataset over time. 
Flares that appear in the SpaceWeatherLive set (and thus have a rank) will be distinguished by red dots. We will look to see if there is any clear trend as to how solar flare starting frequencies varies with the date & time of occurence. 



```python
import matplotlib.pyplot
d = []
f = []
d_top50 = []
f_top50 = []
for i,row in df_nan.iterrows():
    if (row['rank'] == 0):
        #print('yo')
        d.append(datetime.datetime.strptime(row['start_datetime'], "%Y-%m-%d %H:%M:%S"))
        f.append(row['start_frequency'])
    else:
        d_top50.append(datetime.datetime.strptime(row['start_datetime'], "%Y-%m-%d %H:%M:%S"))
        f_top50.append(row['start_frequency'])
#CONVERT to dates usable by matplotlib
dates = matplotlib.dates.date2num(d)
dates_50 = matplotlib.dates.date2num(d_top50)
#matplotlib.pyplot.plot_date(dates, f)
#PLOTTING
#ax = matplotlib.pyplot.axes(xscale='linear',yscale='linear')
#matplotlib.pyplot.locator_params(axis='y',nbins=5)
#ax.xaxis.set_major_locator(matplotlib.pyplot.MaxNLocator(3))
matplotlib.pyplot.plot_date(dates, f, c='b')
matplotlib.pyplot.plot_date(dates_50, f_top50, c ='r')
matplotlib.pyplot.show()
```

The covariance appears to be close to zero here. It looks like the starting frequency of the solar flares is independent of the time at which the solar flares occur.
